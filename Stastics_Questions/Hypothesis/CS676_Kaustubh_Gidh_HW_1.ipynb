```python
# Arithmetic Mean
def arithmetic_mean(list):
  arithmetic_mean = sum(list) / len(list) # Arithmetic Mean = SUM of all values / TOTAL number of values
  return arithmetic_mean
```

```python
# Harmonic Mean
def harmonic_mean(list):
  harmonic_mean = len(list) / sum(1 / x for x in list) # Harmonic Mean = TOTAL number of values / SUM of reciprocal ( Values )
  return harmonic_mean
```

```python
# 1.
print(f"Harmonic Mean of {X} :",harmonic_mean(X))
```

```python
# Geometeric Mean
def geometeric_mean(list):
  multiply = 1 # First we want to calulate multiplication of all values in list ( PIE )
  for x in list:
    multiply = multiply * x # One by One will multiply the values -> 1*4 = 4 -> 4 * 3 = 12
  geometeric_mean = multiply ** (1/len(list)) # Taking nth root of multiplied value -> ( 2 * 3 )**(1/2)
  return geometeric_mean
```

```python
# 1.
print(f"Geometeric Mean of {X} :",geometeric_mean(X))
```

```python
# 2.
print(f"Geometeric Mean of {Y} :",geometeric_mean(Y))
```

```python
# 3.
print(f"Geometeric Mean of {Z} :",geometeric_mean(Z))
```

```python
# Arithmetic Geometeric Mean
def arithmetic_geometeric_mean(list):
  a = arithmetic_mean(list) # Arithmetic Mean
  g = geometeric_mean(list) # Geometeric Mean
  while abs(a-g) > 0.0000001: # Here we are going to run this loop, untill difference between both arithmetic and geometeric is less than 0.0000001
    a , g = arithmetic_mean([a,g]), geometeric_mean([a,g]) # again computing values ( arithmetic mean , geometeric mean )
  return a # Return arithmetic mean value as at this point both value will almost going to be same
```

```python
# 1.
print(f"Arithmetic - Geometeric Mean of {X} :",arithmetic_geometeric_mean(X))
```

```python
# Arithmetic - Harmonic - Geometeric Mean

def arithmetic__harmonic_geometeric_mean(list):
  a = arithmetic_mean(list) # Arithmetic Mean
  h = harmonic_mean(list) # Harmonic Mean
  g = geometeric_mean(list) # Geometeric Mean

  while max(abs(a-h),abs(a-g),abs(g-h)) > 0.0000001: # Here we are going to run this loop, until difference between is less than this value
    temp_a = (a+h+g) / 3 # again computing mean and storing as a new (temp) mean
    temp_g = (a*h*g)**(1/3)
    h = 3/ (1/a + 1/h + 1/g) # Harmonic Mean
    a = temp_a
    g = temp_g
  return (a + g + h)/3
```

```python
# 1.
print(f"Arithmetic - Harmonic - Geometeric Mean of {X} :",arithmetic__harmonic_geometeric_mean(X))
```

```python
# Median

# even values - 4 6 3 7 3 4 5 8
#                     | |

# odd values - 4 5 2 6 7
#                  |

def median(list):
  data = sorted(list) # First we are going to sort the list
  n = len(data) # No. of values in list
  mid = n//2
  if n %2 == 0 : # If no. of values are even
    median = (data[mid] + data[mid-1])/2
  else: # If no. of values are odd
    median = data[mid]
  return median
```

```python
# question a

# arithmetic mean weighted is -> Sum (B * f) / Sum (f)
def arithmetic_mean_weighted(B,f):
  total = 0 # intially both total and count will be zero
  count = 0

  for i in range (len(B)): # run loop till len of B
    total = total + B[i] * f[i]
    count = count + f[i]

  return total / count
```

```python
# question b

# Mode
def mode_w(B, f):
    max_frequent = f[0] # Variable for storing most frequent element
    mode_value = B[0]

    for i in range(1, len(f)):
        if f[i] > max_frequent:
            max_frequent = f[i]
            mode_value = B[i]

    return mode_value # the number who has most number of frequency will be returned
```

```python
# question c
def harmonic_mean_weighted(B, f):
    numerator = 0
    denominator = 0

    for i in range(len(B)):
        numerator = numerator + f[i]
        denominator = denominator + (f[i] / B[i])

    return numerator / denominator
```

```python
harmonic_mean_weighted(B,f)
```

```python
# question d
def geometeric_mean_weighted(B, f):
    log_sum = 0
    count = 0

    for i in range(len(B)):
        log_sum = log_sum + f[i] * math.log(B[i])
        count = count + f[i]

    return math.exp(log_sum / count)
```

```python
geometeric_mean_weighted(B,f)
```

```python
# question e
def arithmetic_geometeric_mean_weighted(B,f):
  a = arithmetic_mean_weighted(B,f) # Arithmetic Mean
  g = geometeric_mean_weighted(B,f) # Geometeric Mean
  while abs(a-g) > 0.0000001:
     # Here we are going to run this loop, untill difference between both arithmetic and geometeric is less than 0.0000001
    new_a = (a + g) / 2 # arithemtic mean
    new_g = (a * g)**(1/2) # geometeric
    a = new_a
    g = new_g
  return a
```
